```python
ouvrir=open(path)
lecture=ouvrir.readlines()
print(lecture)
```

```python
identifiants=[objet.strip('\n').split(';')[0] for objet in lecture]
Point=[objet.strip('\n').split(';')[1] for objet in lecture]
usage=[objet.strip('\n').split(';')[2] for objet in lecture]
fonctionel=[objet.strip('\n').split(';')[3] for objet in lecture]
Latitude=[objet.strip('\n').split(';')[4] for objet in lecture]
Longitude=[objet.strip('\n').split(';')[5] for objet in lecture]
Altitude=[objet.strip('\n').split(';')[6] for objet in lecture]
E=[objet.strip('\n').split(';')[7] for objet in lecture]
N=[objet.strip('\n').split(';')[8] for objet in lecture]
Z=[objet.strip('\n').split(';')[9] for objet in lecture]
Height=[objet.strip('\n').split(';')[10] for objet in lecture]
```

```python
dicc={}
dicc['idd']=identifiants
dicc['pt']=Point
dicc['usag']=usage
dicc['fonct']=fonctionel
dicc['lat']=Latitude
dicc['lon']=Longitude
dicc['alt']=Altitude
dicc['e']=E
dicc['n']=N
dicc['z']=Z
dicc['h']=Height
print(dicc)
```

```python
import json
with open ("dicc.json",'w') as data_json:
    json.dump(dicc,data_json)
    data_json.close()
```
