Gestionnaire de tâches : Créez un programme qui permet à l'utilisateur de gérer une liste de tâches. Il devrait être capable d'ajouter de nouvelles tâches, de marquer des tâches comme terminées et de supprimer des tâches de la liste.



```python
def afficher_taches(mes_taches):
    if len(mes_taches) == 0:
        print("La liste des tâches est vide.")
    else:
        print("Liste des tâches disponibles:")
        for index, tache in enumerate(mes_taches):
            print(f'{index+1}: {tache}')

def gestionnaire_taches():
    mes_taches = ['apprendre', 'lessiver', 'surveiller_gazon', 'aller_bank']
    
    print("\n\nGESTIONNAIRE DES TACHES\n\n".upper())
    
    while True: 
        afficher_taches(mes_taches)
        
        ajouter_tache = input('Voulez-vous ajouter une tâche ? (oui/non): ')
        
        if ajouter_tache.lower() == 'oui':
            nouvelle_tache = input("Ajouter une tâche à la liste: ")
            mes_taches.append(nouvelle_tache)
            print('La tâche a été ajoutée à la liste.')
        
        supprimer_tache = input('Voulez-vous supprimer une tâche ? (oui/non): ')
        
        if supprimer_tache.lower() == 'oui':
            tache_supprimer = input("Entrez le nom de la tâche à supprimer: ")
            mes_taches.remove(tache_supprimer)
            print('La tâche a été supprimée avec succès.')
        
        tache_terminee = input('Voulez-vous exclure une tâche ? (oui/non): ')
        
        if tache_terminee.lower() == 'oui':
            numero_tache = int(input("Entrez le numéro de la tâche à exclure: "))
            
            if 1 <= numero_tache <= len(mes_taches):
                mes_taches.pop(numero_tache - 1)
                print('La tâche a été exclue avec succès.')
            else:
                print('Numéro de tâche invalide.')
        
        fermer = input("Voulez-vous fermer le gestionnaire de tâches ? (oui/non): ")
        
        if fermer.lower() == 'oui':
            break

gestionnaire_taches()
```

```python
# Fonction pour afficher la liste de tâches
def afficher_taches(taches):
    if len(taches) == 0:
        print("La liste de tâches est vide.")
    else:
        print("Liste de tâches :")
        for i, tache in enumerate(taches):
            print(f"{i+1}. {tache}")

# Fonction principale
def gestionnaire_taches():
    taches = []
    while True:
        print("\n===== Gestionnaire de tâches =====")
        print("1. Afficher la liste de tâches")
        print("2. Ajouter une tâche")
        print("3. Marquer une tâche comme terminée")
        print("4. Supprimer une tâche")
        print("5. Quitter le programme")

        choix = input("Choisissez une option (1-5) : ")
        if choix == "1":
            afficher_taches(taches)
        elif choix == "2":
            nouvelle_tache = input("Entrez la nouvelle tâche : ")
            taches.append(nouvelle_tache)
            print("La tâche a été ajoutée avec succès.")
        elif choix == "3":
            afficher_taches(taches)
            num_tache = int(input("Entrez le numéro de la tâche terminée : "))
            if num_tache >= 1 and num_tache <= len(taches):
                taches.pop(num_tache - 1)
                print("La tâche a été marquée comme terminée.")
            else:
                print("Numéro de tâche invalide.")
        elif choix == "4":
            afficher_taches(taches)
            num_tache = int(input("Entrez le numéro de la tâche à supprimer : "))
            if num_tache >= 1 and num_tache <= len(taches):
                taches.pop(num_tache - 1)
                print("La tâche a été supprimée de la liste.")
            else:
                print("Numéro de tâche invalide.")
        elif choix == "5":
            print("Merci d'avoir utilisé le gestionnaire de tâches. Au revoir !")
            break
        else:
            print("Option invalide. Veuillez sélectionner une option valide (1-5).")

# Appel de la fonction principale
gestionnaire_taches()
```
