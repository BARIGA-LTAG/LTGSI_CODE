```python
def transfert():
    option = input("entrez le numéro d'option: ")
    if option == "1":
        print("entre 0 pour passer au favoris ou cash")
        montant = int(input("entrez le montant: "))
        les_frais=0.08*montant
        les_frais_2=0.02*montant
        if montant != 0:
            numero_destinataire = int(input("numéro du destinataire: "))
            frais=["oui","non"]
            for index, value in enumerate(frais):
                print(f"ajouter des frais de retrait pour le destinataire ? \n{index+1}. {value}")
            option_A = input()
            if option_A == "1":
                print(f"vous voulez envoyer {montant +les_frais} au {numero_destinataire}, frais: {les_frais_2} FCFA")
            else:
                print(f"vous voulez envoyer {montant} au {numero_destinataire}, frais: {les_frais_2} FCFA")
        else:
            autre_choix = ["Pour accéder aux favoris","TCsh:sans numéro"]
            for index, code in enumerate(autre_choix):
                print(f"{index+1} {code}")
                option_B = input()
                if option_B == "1":
                    code_secret = input("entrez votre code secret: ")
                    contact_favoris = []
                    if len(contact_favoris) == 0:
                        print("aucun contact enregistré.")
                else:
                    prenom = input("entrez le prénom du destinataire: ")
                    nom = input("entrez le nom du destinataire: ")
                    print(f"vous allez effectuer un envoi cash de {montant} FCFA pour {prenom} {nom}\n frais: __ FCFA, Entrez votre code secret (0 pour annuler)")
transfert()
```
